```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def sigmoid(x):
  return 1 / (1 + np.exp(-x))
```

```python
v = np.linspace(-10, 10, 1000)
a = sigmoid(v)

plt.title('sigmoid function')
plt.plot(a)
plt.show()
```

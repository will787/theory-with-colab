```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def sigmoid(x):
  return 1 / (1 + np.exp(-x))
```

```python
v = np.linspace(-10, 10, 1000)
a = sigmoid(v)

plt.title('sigmoid function')
plt.plot(a)
plt.show()
```

```python
insumos = np.linspace(-20, 20, 40)
print(insumos)
pt = []


for i in insumos:
  pt.append(10/i)


print(pt)

plt.plot(insumos,pt)
plt.show()
```
